```python
years = range(2000, 2009)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'oranges']);
```

```python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'oranges'])
```

To get whole set of matplotlib markers:
https://matplotlib.org/stable/api/markers_api.html

```python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'oranges']);
```

to see the full list of arguements that plt() can take: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

The fmt arguement provides a shorthand for specifying the line style, marker, and line color. It can be provided as the third arguement to plt.plot()

```python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'oranges']);
```

```python
plt.plot(years, oranges, 'or')
plt.title("Yields of oranges (tons per hectares)");
```

```python
plt.figure(figsize=(12,6))

plt.plot(years, oranges, 'or')
plt.title("Yields of oranges (tons per hectares)");
```

Scatter Plot

```python
#Load data into a pandas dataframe
flower_df = sns.load_dataset("iris")  #using an inbuilt dataset from seaborn
```

```python
flower_df
```

```python
flower_df.species.unique()
```

```python
plt.plot(flower_df.sepal_length, flower_df.sepal_width);
```

```python
sns.scatterplot(data=flower_df, x='sepal_length', y='sepal_width')
```

```python
plt.figure(figsize=(12,6))
plt.title("Sepal Dimensions")
sns.scatterplot(data=flower_df, x='sepal_length', y='sepal_width', hue='species', s=100)
```

```python
flower_df = sns.load_dataset("iris")
```
