```python
import matplotlib.pyplot as plt
import numpy as np
```

Line Plot

```python
x = np.array([0,50])
y = np.array([0,50])
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.array([1,2,5,7])
y = np.array([3,8,4,10])

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,linestyle="dotted")
plt.show()
```

```python
plt.plot(x,y, marker='*', linestyle='dashed')
```

```python
plt.plot(x,y,linestyle="dotted")
plt.grid()
plt.show()
```

```python
plt.plot(x,y,linestyle="dotted")
plt.xlabel("This is a x-axis")
plt.ylabel("This is y-axis")
plt.title("Line plot example graph")
plt.grid()

plt.show()
```

Sub-Plot

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,3,1)
plt.plot(x,y)

a = np.array([0,1,2,3])
b = np.array([10,20,30,40])

plt.subplot(1,3,2)
plt.plot(a,b)
plt.show()
```

Scatter Plots

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,101,86,89,94,78,77,85,86,101])
plt.scatter(x,y)
plt.show()
```
